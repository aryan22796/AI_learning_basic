# Day 4 Basic Assignment Notebook

This notebook is the beginner-friendly version of the Day 4 assignment.

We will:
1. Load the healthcare dataset.
2. Inspect the table.
3. Calculate simple summary statistics.
4. Create a new column called `cost_per_visit`.
5. Save a report file for later use.

```python
# ------------------------------------------------------------------
# Cell 1: Import libraries and prepare the dataset path
# ------------------------------------------------------------------
# pandas is used to read and analyze table-style data.
import pandas as pd

# pathlib.Path helps us locate files safely, even if the notebook runs
# from different folders inside the project.
from pathlib import Path

# Try the path from the workspace root first.
DATA_PATH = Path("Day4_Assignment/data/healthcare_assignment_dataset.csv")

# If the notebook is opened from inside Day4_Assignment/basic, use the
# folder-relative path instead.
if not DATA_PATH.exists():
    DATA_PATH = Path("../data/healthcare_assignment_dataset.csv")

# Print the final path so we know which file is being used.
print("Dataset path:", DATA_PATH)
```

```python
# ------------------------------------------------------------------
# Cell 2: Load the CSV into a DataFrame
# ------------------------------------------------------------------
# pd.read_csv() reads the CSV file into a table object called a DataFrame.
df = pd.read_csv(DATA_PATH)

# Show a short preview of the dataset.
print("Basic assignment started.")
print(df.head())

# Print the number of rows and columns.
print("Shape:", df.shape)
```

```python
# ------------------------------------------------------------------
# Cell 3: Calculate simple summary statistics
# ------------------------------------------------------------------
# Compute the average age of the patients in the dataset.
avg_age = df["age"].mean()

# Compute the average total cost across all patients.
avg_cost = df["total_cost"].mean()

# Print both values to understand the dataset quickly.
print("Average age:", round(avg_age, 2))
print("Average total cost:", round(avg_cost, 2))
```

```python
# ------------------------------------------------------------------
# Cell 4: Create a new business metric
# ------------------------------------------------------------------
# cost_per_visit tells us how much each patient spends per visit.
# This is useful for comparing patients and identifying expensive cases.
df["cost_per_visit"] = df["total_cost"] / df["visits"]

# Show only the useful columns to keep the output readable.
print(df[["patient_id", "condition", "visits", "total_cost", "cost_per_visit"]].head())
```

```python
# ------------------------------------------------------------------
# Cell 5: Save the updated dataset as a report
# ------------------------------------------------------------------
# Create the output folder if it does not exist yet.
OUTPUT_PATH = Path("Day4_Assignment/output/basic_report.csv")
OUTPUT_PATH.parent.mkdir(exist_ok=True)

# Save the updated DataFrame to CSV.
df.to_csv(OUTPUT_PATH, index=False)

# Print where the report was saved.
print("Basic report saved to:", OUTPUT_PATH)
```
