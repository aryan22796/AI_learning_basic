This notebook builds on the beginner assignment.

We will practice:
- writing reusable functions,
- filtering high-cost patients,
- summarizing by department, and
- using a simple class to model a patient summary.

```python
# ------------------------------------------------------------------
# Cell 1: Load the dataset using a reusable function
# ------------------------------------------------------------------
# Import pandas to work with tabular healthcare data.
import pandas as pd

# Import Path for safe file handling.
from pathlib import Path

# Define a function that loads the CSV and returns a DataFrame.
def load_dataset(file_path: str) -> pd.DataFrame:
    # Convert the string into a Path object for reliability.
    path = Path(file_path)
    # If the path does not exist from the root folder, try the relative path.
    if not path.exists():
        path = Path("Day4_Assignment/data/healthcare_assignment_dataset.csv")
    # Read the CSV file and return the table.
    return pd.read_csv(path)

# Load the dataset using the helper function.
df = load_dataset("../data/healthcare_assignment_dataset.csv")
print("Loaded rows:", df.shape)
```

```python
# ------------------------------------------------------------------
# Cell 2: Add the cost_per_visit field
# ------------------------------------------------------------------
# Define a function that adds cost per visit to the DataFrame.
def add_cost_per_visit(data_frame: pd.DataFrame) -> pd.DataFrame:
    # Make a copy so the original data is not changed by mistake.
    working_frame = data_frame.copy()
    # Create the new business metric.
    working_frame["cost_per_visit"] = working_frame["total_cost"] / working_frame["visits"]
    # Return the updated table.
    return working_frame

# Use the function and preview the result.
df = add_cost_per_visit(df)
print(df[["patient_id", "condition", "total_cost", "visits", "cost_per_visit"]].head())
```

```python
# ------------------------------------------------------------------
# Cell 3: Find high-cost patients
# ------------------------------------------------------------------
# Define a function that filters rows above a chosen limit.
def find_high_cost_patients(data_frame: pd.DataFrame, limit: float) -> pd.DataFrame:
    # Filter all rows whose total_cost is greater than the limit.
    high_cost = data_frame[data_frame["total_cost"] > limit]
    # Return the subset of patients that match.
    return high_cost

# Find all patients above 2000 cost and show the first few.
high_cost_df = find_high_cost_patients(df, 2000)
print("High-cost patients:")
print(high_cost_df[["patient_id", "department", "condition", "total_cost", "cost_per_visit"]].head(10))
```

```python
# ------------------------------------------------------------------
# Cell 4: Summarize by department
# ------------------------------------------------------------------
# Group patients by department and compute the mean total cost.
summary_by_department = df.groupby("department")["total_cost"].mean().sort_values(ascending=False)

# Print the summary table.
print("Average cost by department:")
print(summary_by_department)
```

```python
# ------------------------------------------------------------------
# Cell 5: Use a simple class for patient summary
# ------------------------------------------------------------------
# Define a class that stores a patient summary.
class PatientSummary:
    # The constructor stores the most useful fields.
    def __init__(self, patient_id: int, condition: str, total_cost: float) -> None:
        self.patient_id = patient_id
        self.condition = condition
        self.total_cost = total_cost

    # This method tells us if the patient is expensive.
    def is_expensive(self) -> bool:
        return self.total_cost > 2500

# Create one object from the class.
patient = PatientSummary(1009, "Hypertension", 2100.25)

# Print the object details and the result of the method.
print("Patient summary object:")
print("Patient ID:", patient.patient_id)
print("Condition:", patient.condition)
print("Is expensive?", patient.is_expensive())
```

```python
# ------------------------------------------------------------------
# Cell 6: Save the summary output
# ------------------------------------------------------------------
# Create the output folder, if needed.
OUTPUT_PATH = Path("Day4_Assignment/output/intermediate_summary.csv")
OUTPUT_PATH.parent.mkdir(exist_ok=True)

# Save the department summary to CSV.
summary_by_department.to_csv(OUTPUT_PATH)
print("Intermediate summary saved to:", OUTPUT_PATH)
```
