## Part 1: Advanced NumPy

### 1.1 Linear Algebra & Matrix Operations

```python
import numpy as np
from numpy.linalg import eig, svd, solve, det, inv
import matplotlib.pyplot as plt

# Matrix operations
A = np.array([[4, 2], [1, 3]])
B = np.array([[1, 0], [0, 2]])

print("Matrix A:\n", A)
print("\nMatrix multiplication (A @ B):\n", A @ B)
print("\nElement-wise multiplication (A * B):\n", A * B)

# Determinant and inverse
print("\nDeterminant of A:", det(A))
print("Inverse of A:\n", inv(A))

# Eigenvalues and eigenvectors
eigenvalues, eigenvectors = eig(A)
print("\nEigenvalues:", eigenvalues)
print("Eigenvectors:\n", eigenvectors)

# Singular Value Decomposition
U, S, Vt = svd(A)
print("\nSingular Values:", S)
print("U matrix shape:", U.shape, "Vt matrix shape:", Vt.shape)
```

### 1.2 Advanced Array Operations & Structured Arrays

```python
# Structured arrays with named fields
dt = np.dtype([('name', 'U10'), ('age', 'i4'), ('salary', 'f8')])
employees = np.array([('Alice', 30, 75000), ('Bob', 35, 85000), ('Charlie', 28, 65000)], dtype=dt)

print("Structured array:")
print(employees)
print("\nNames:", employees['name'])
print("Average salary:", employees['salary'].mean())

# Broadcasting with different dimensions
data = np.arange(24).reshape(2, 3, 4)
bias = np.array([10, 20, 30, 40])
print("\nData shape:", data.shape)
print("Bias shape:", bias.shape)
print("Data + bias shape:", (data + bias).shape)

# Fancy indexing
arr = np.arange(20).reshape(4, 5)
indices = np.array([0, 2])
print("\nOriginal array:\n", arr)
print("\nRows [0, 2]:\n", arr[indices])
print("\nElements at (row, col): (0,2) and (2,4):", arr[np.array([0, 2]), np.array([2, 4])])
```

### 1.3 Advanced Numerical Functions & FFT

```python
# Polynomial operations
p = np.poly1d([1, 2, 1])  # 1*x^2 + 2*x + 1
print("Polynomial:", p)
print("Polynomial roots:", np.roots([1, 2, 1]))
print("Evaluate at x=5:", p(5))

# FFT (Fast Fourier Transform) for signal processing
t = np.linspace(0, 2*np.pi, 100)
signal = np.sin(t) + 0.5*np.sin(5*t)
fft_result = np.fft.fft(signal)
frequencies = np.fft.fftfreq(len(signal))

print("\nSignal shape:", signal.shape)
print("FFT magnitude (first 10):", np.abs(fft_result[:10]))
print("Dominant frequency index:", np.argmax(np.abs(fft_result[:50])))

# Convolution (useful for smoothing, filtering)
data = np.array([1, 2, 3, 4, 5, 4, 3, 2, 1])
kernel = np.array([0.2, 0.6, 0.2])  # Simple smoothing kernel
smoothed = np.convolve(data, kernel, mode='same')
print("\nOriginal signal:", data)
print("Smoothed signal:", np.round(smoothed, 2))
```

### 1.4 Random Number Generation & Sampling

```python
np.random.seed(42)

# Various distributions
normal = np.random.normal(loc=100, scale=15, size=1000)
uniform = np.random.uniform(low=0, high=100, size=1000)
exponential = np.random.exponential(scale=2, size=1000)
poisson = np.random.poisson(lam=5, size=1000)

print("Normal distribution - mean: {:.2f}, std: {:.2f}".format(normal.mean(), normal.std()))
print("Uniform distribution - min: {:.2f}, max: {:.2f}".format(uniform.min(), uniform.max()))
print("Poisson distribution - mean: {:.2f}".format(poisson.mean()))

# Sampling without replacement
population = np.arange(1, 101)
sample = np.random.choice(population, size=20, replace=False)
print("\nSample of 20 from 1-100:", sorted(sample))

# Weighted sampling
weights = np.array([0.1, 0.3, 0.4, 0.2])
categories = np.array(['A', 'B', 'C', 'D'])
weighted_sample = np.random.choice(categories, size=1000, p=weights)
unique, counts = np.unique(weighted_sample, return_counts=True)
print("\nWeighted sampling distribution:")
for cat, count in zip(unique, counts):
    print(f"  {cat}: {count/1000:.1%}")
```

## Part 2: Advanced Pandas

### 2.1 MultiIndex (Hierarchical Indexing)

```python
import pandas as pd

# Create hierarchical index
arrays = [
    ['Product A', 'Product A', 'Product B', 'Product B'],
    ['Q1', 'Q2', 'Q1', 'Q2']
]
index = pd.MultiIndex.from_arrays(arrays, names=['Product', 'Quarter'])
sales = pd.DataFrame({'Revenue': [100000, 120000, 85000, 95000], 'Units': [500, 600, 450, 500]}, index=index)

print("MultiIndex DataFrame:")
print(sales)

print("\nAccess Product A:")
print(sales.loc['Product A'])

print("\nAccess Product A, Q1:")
print(sales.loc[('Product A', 'Q1')])

# Unstack and stack operations
unstacked = sales.unstack()
print("\nUnstacked:")
print(unstacked)

stacked = unstacked.stack()
print("\nStacked back:")
print(stacked)
```

### 2.2 Time Series Analysis

```python
# Create time series data
dates = pd.date_range('2024-01-01', periods=365, freq='D')
np.random.seed(42)
values = 100 + np.cumsum(np.random.randn(365) * 2)
ts = pd.DataFrame({'price': values}, index=dates)

print("Time Series:")
print(ts.head(10))

# Rolling windows
ts['ma_7'] = ts['price'].rolling(window=7).mean()
ts['ma_30'] = ts['price'].rolling(window=30).mean()
ts['volatility'] = ts['price'].rolling(window=20).std()

print("\nWith rolling statistics:")
print(ts.tail(10))

# Expanding windows
ts['cummax'] = ts['price'].expanding().max()
ts['cumsum'] = ts['price'].expanding().sum()

# Resampling
weekly_agg = ts['price'].resample('W').agg(['mean', 'min', 'max'])
print("\nWeekly aggregation:")
print(weekly_agg.head())

# Shifting and percentage change
ts['price_lag_1'] = ts['price'].shift(1)
ts['price_change'] = ts['price'].pct_change()
print("\nPrice lag and percentage change:")
print(ts[['price', 'price_lag_1', 'price_change']].head())
```

### 2.3 Advanced Groupby with Multiple Aggregations

```python
# Create rich dataset
np.random.seed(42)
data = pd.DataFrame({
    'region': np.repeat(['North', 'South', 'East', 'West'], 50),
    'product': np.tile(np.repeat(['A', 'B'], 25), 4),
    'sales': np.random.randint(100, 1000, 200),
    'quantity': np.random.randint(1, 50, 200),
    'date': pd.date_range('2024-01-01', periods=200, freq='D')
})

# Complex aggregation
agg_dict = {
    'sales': ['sum', 'mean', 'std', ('q75', lambda x: x.quantile(0.75))],
    'quantity': ['sum', 'min', 'max'],
}

grouped = data.groupby(['region', 'product']).agg(agg_dict)
print("Multi-level aggregation:")
print(grouped)

# Named aggregations (cleaner)
summary = data.groupby('region').agg(
    total_sales=('sales', 'sum'),
    avg_sales=('sales', 'mean'),
    sale_std=('sales', 'std'),
    total_qty=('quantity', 'sum'),
    avg_price=('sales', lambda x: x.mean() / data.loc[x.index, 'quantity'].mean())
).round(2)
print("\nNamed aggregations:")
print(summary)
```

### 2.4 String Operations & Categorical Data

```python
# String operations
names = pd.Series(['John Smith', 'JANE DOE', 'bob johnson', 'ALICE WILLIAMS'])
print("Original names:", names.tolist())
print("Lowercase:", names.str.lower().tolist())
print("Uppercase:", names.str.upper().tolist())
print("Title case:", names.str.title().tolist())
print("Length:", names.str.len().tolist())
print("Contains 'John':", names.str.contains('John').tolist())
print("Extract first name:", names.str.split().str[0].tolist())

# Categorical data optimization
df = pd.DataFrame({
    'category': ['A', 'B', 'A', 'C', 'B', 'A'] * 100,
    'value': np.random.randn(600)
})

# Before: object dtype
print("\nBefore categorical:")
print(f"Memory: {df['category'].memory_usage(deep=True)} bytes")

# After: categorical
df['category'] = df['category'].astype('category')
print("\nAfter categorical:")
print(f"Memory: {df['category'].memory_usage(deep=True)} bytes")
print(f"Categories: {df['category'].cat.categories.tolist()}")
```

## Part 3: Advanced EDA Techniques

### 3.1 Statistical Profiling

```python
from scipy import stats

# Create dataset
np.random.seed(42)
df = pd.DataFrame({
    'age': np.random.normal(45, 15, 1000),
    'income': np.random.gamma(shape=2, scale=20000, size=1000),
    'visits': np.random.poisson(lam=5, size=1000),
    'score': np.random.beta(a=2, b=5, size=1000) * 100
})

# Statistical summary
print("Statistical Profiling:")
for col in df.columns:
    skewness = stats.skew(df[col])
    kurtosis = stats.kurtosis(df[col])
    norm_stat, norm_p = stats.normaltest(df[col])
    print(f"\n{col}:")
    print(f"  Skewness: {skewness:.3f}")
    print(f"  Kurtosis: {kurtosis:.3f}")
    print(f"  Normal Test p-value: {norm_p:.6f}")
    print(f"  Is normal? {norm_p > 0.05}")
```

### 3.2 Correlation Analysis & Feature Relationships

```python
# Correlation matrices
corr_pearson = df.corr(method='pearson')
corr_spearman = df.corr(method='spearman')
corr_kendall = df.corr(method='kendall')

print("Pearson Correlation:")
print(corr_pearson.round(3))

# P-values for correlations
from scipy.stats import pearsonr

print("\nCorrelation with p-values:")
for i, col1 in enumerate(df.columns):
    for col2 in df.columns[i+1:]:
        corr, pval = pearsonr(df[col1], df[col2])
        significance = '***' if pval < 0.001 else '**' if pval < 0.01 else '*' if pval < 0.05 else 'ns'
        print(f"{col1} vs {col2}: {corr:.3f} (p={pval:.4f}) {significance}")
```

### 3.3 Outlier Detection

```python
# Add some outliers
df_with_outliers = df.copy()
df_with_outliers.loc[5, 'age'] = 150
df_with_outliers.loc[10, 'income'] = 500000
df_with_outliers.loc[15, 'visits'] = 100

# IQR Method
def detect_outliers_iqr(data, col, multiplier=1.5):
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - multiplier * IQR
    upper_bound = Q3 + multiplier * IQR
    return (data[col] < lower_bound) | (data[col] > upper_bound)

# Z-Score Method
def detect_outliers_zscore(data, col, threshold=3):
    z_scores = np.abs(stats.zscore(data[col]))
    return z_scores > threshold

print("Outlier Detection:")
for col in df_with_outliers.columns:
    iqr_outliers = detect_outliers_iqr(df_with_outliers, col).sum()
    z_outliers = detect_outliers_zscore(df_with_outliers, col).sum()
    print(f"\n{col}:")
    print(f"  IQR method outliers: {iqr_outliers}")
    print(f"  Z-score method outliers: {z_outliers}")
```

### 3.4 Feature Engineering & Data Quality Assessment

```python
# Create dataset with data quality issues
np.random.seed(42)
df_messy = pd.DataFrame({
    'user_id': range(1, 101),
    'age': np.random.randint(18, 80, 100),
    'income': np.random.randint(20000, 150000, 100),
    'score': np.random.uniform(0, 100, 100),
    'is_active': np.random.choice([True, False, None], 100),
    'signup_date': pd.date_range('2023-01-01', periods=100, freq='D')
})

# Add some missing values
df_messy.loc[np.random.choice(100, 10, replace=False), 'income'] = np.nan
df_messy.loc[np.random.choice(100, 5, replace=False), 'score'] = np.nan

# Data Quality Assessment
print("Data Quality Report:")
print("\nMissing Values:")
missing = df_messy.isnull().sum()
missing_pct = (missing / len(df_messy) * 100).round(2)
for col in df_messy.columns:
    if missing[col] > 0:
        print(f"  {col}: {missing[col]} ({missing_pct[col]}%)")

print("\nData Types:")
print(df_messy.dtypes)

print("\nDuplicate Rows:", df_messy.duplicated().sum())

# Feature Engineering
df_messy['age_group'] = pd.cut(df_messy['age'], bins=[0, 30, 45, 60, 100], 
                                 labels=['Young', 'Middle', 'Senior', 'Elderly'])
df_messy['income_category'] = pd.qcut(df_messy['income'].dropna(), q=3, 
                                        labels=['Low', 'Medium', 'High'])
df_messy['days_since_signup'] = (pd.Timestamp.now() - df_messy['signup_date']).dt.days
df_messy['score_normalized'] = (df_messy['score'] - df_messy['score'].mean()) / df_messy['score'].std()

print("\nFeature-engineered DataFrame:")
print(df_messy.head())
```
