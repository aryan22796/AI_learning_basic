### Reading and Writing Files
```python
sample_path = "sample_text.txt"
with open(sample_path, "w", encoding="utf-8") as f:
    f.write("Line 1\nLine 2\n")

with open(sample_path, "r", encoding="utf-8") as f:
    content = f.read()
print(content)
```

This example writes text to a file and reads it back safely using a context manager.


````python
sample_path = "sample_text.txt"
with open(sample_path, "w", encoding="utf-8") as f:
    f.write("Line 1\nLine 2\n")

with open(sample_path, "r", encoding="utf-8") as f:
    content = f.read()
print(content)
````

### List Comprehensions
```python
numbers = [10, 20, 30, 40]
multiplied = [value * 2 for value in numbers if value > 15]
print(multiplied)  # [40, 60, 80]
```

List comprehensions are a concise way to transform lists with filtering in one line.


````python
numbers = [10, 20, 30, 40]
multiplied = [value * 2 for value in numbers if value > 15]
print(multiplied)  # [40, 60, 80]
````

## Guided Lab: Customer JSON Parser
1. Create a `customers.json` file with sample records.
2. Load the JSON file.
3. Filter active customers.
4. Format the output.
5. Save the results to a new JSON file.
6. Handle missing files and invalid JSON.


````python
import json

sample_data = [
    {"id": 101, "name": "Aisha Khan", "status": "active", "email": "aisha@example.com"},
    {"id": 102, "name": "Leon Wu", "status": "inactive", "email": "leon@example.com"},
    {"id": 103, "name": "Sara Patel", "status": "active", "email": "sara@example.com"}
]

with open("lab/customers.json", "w", encoding="utf-8") as f:
    json.dump(sample_data, f, indent=2)

with open("lab/customers.json", "r", encoding="utf-8") as f:
    customers = json.load(f)

active_customers = [
    {
        "customerId": record["id"],
        "customerName": record["name"],
        "email": record["email"],
        "customerStatus": record["status"]
    }
    for record in customers
    if record.get("status") == "active"
]

with open("lab/active_customers.json", "w", encoding="utf-8") as f:
    json.dump(active_customers, f, indent=2)

print(f"Saved {len(active_customers)} active customers to lab/active_customers.json")
````
