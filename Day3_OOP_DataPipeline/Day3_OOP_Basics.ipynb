This file uses a real healthcare example instead of abstract theory. We will see how OOP helps us model patient records in a clean way.

## Why this example is practical
We will use the healthcare dataset to show:
- what a class is
- what `self` means
- what `__init__()` does
- how inheritance works
- why type hints are useful

The goal is simple: learn OOP by solving a real data problem, not by memorizing definitions only.

```python
import pandas as pd            # Import pandas to read and work with the healthcare CSV file
from pathlib import Path        # Import Path to locate the dataset file safely

# Step 1: Find the CSV file in the project
csv_path = Path('data/healthcare.csv')          # Try the normal folder path first
if not csv_path.exists():                       # If that path does not exist
    csv_path = Path('Day3_OOP_DataPipeline/data/healthcare.csv')  # use the Day 3 folder path

# Step 2: Read the CSV and take the first patient row
df = pd.read_csv(csv_path)                      # Load the healthcare table into a DataFrame
first_row = df.iloc[0].to_dict()                # Take the first row and convert it to a dictionary
print('First patient row from the dataset:')    # Print a heading
print(first_row)                                # Show the first patient record

# Step 3: Create a class that represents one patient record
class PatientRecord:
    def __init__(self, patient_id: int, age: int, total_cost: float, visits: int):
        # __init__ runs automatically when we create an object
        # self means 'this object' so its values stay inside this patient object
        self.patient_id = patient_id            # Save patient ID in this object
        self.age = age                          # Save age in this object
        self.total_cost = total_cost            # Save total cost in this object
        self.visits = visits                    # Save number of visits in this object

    def cost_per_visit(self) -> float:
        # This method calculates the average cost for one visit
        return self.total_cost / self.visits    # Divide total cost by visit count

# Step 4: Create one real object from the class blueprint
patient_one = PatientRecord(
    patient_id=first_row['patient_id'],         # Use the real patient id from the dataset
    age=first_row['age'],                       # Use the real age from the dataset
    total_cost=first_row['total_cost'],         # Use the real total cost
    visits=first_row['visits'],                 # Use the real number of visits
)

# Step 5: Print the result
print('\nCost per visit for patient 1001:', round(patient_one.cost_per_visit(), 2))
# This shows how OOP makes it easy to work with one patient record at a time
```

## What is `self`? (definition)
`self` is a special name that refers to the current object being created or used.

## Why do we need `self`? (example)
When we write `patient_one = PatientRecord(...)`, Python creates one object. Inside the class, `self` tells Python: 'store these values inside this exact object only.'

Example:
- `self.patient_id = 1001`
- `self.age = 45`

Without `self`, Python would not know which patient object we are talking about.

```python
# ------------------------------------------------------------
# Inheritance + Polymorphism example (completed)
# ------------------------------------------------------------

class Patient:
    # Parent class: common behavior for all patient types
    def __init__(self, patient_id: int, age: int):
        self.patient_id = patient_id   # store patient ID
        self.age = age                 # store patient age

    def describe(self) -> str:
        # Normal description used by all patients
        return f'Patient {self.patient_id} is {self.age} years old.'


class HighRiskPatient(Patient):
    # Child class: inherits from Patient and adds extra information
    def __init__(self, patient_id: int, age: int, condition: str):
        super().__init__(patient_id, age)  # call the parent constructor
        self.condition = condition          # add condition-specific detail

    def describe(self) -> str:
        # Same method name as parent, but different behavior
        return f'High-risk patient {self.patient_id} has {self.condition}.'


class FollowUpPatient(Patient):
    # Another child class that shows polymorphism in a second way
    def __init__(self, patient_id: int, age: int, department: str):
        super().__init__(patient_id, age)
        self.department = department

    def describe(self) -> str:
        return f'Follow-up patient {self.patient_id} is in {self.department}.'


# Create different objects
patient_a = Patient(1001, 45)
patient_b = HighRiskPatient(1008, 63, 'Diabetes')
patient_c = FollowUpPatient(1015, 50, 'Pulmonology')

# Store them in one list
patients = [patient_a, patient_b, patient_c]

# Loop through all objects and call the same method name: describe()
for p in patients:
    print(p.describe())

# Why this is important:
# Even though all objects are different, we can use the same method name
# and Python will call the correct version for each object.
# This is polymorphism.

# Why inheritance matters:
# HighRiskPatient and FollowUpPatient reuse Patient code
# instead of rewriting the same basic logic again.
```

## Patient 1001: where inheritance and polymorphism appear

This example uses the real patient record with ID 1001.

1. Inheritance:
   - `HighRiskPatient` is created from `Patient`.
   - It reuses the basic patient fields like `patient_id` and `age`.

2. Polymorphism:
   - Both `Patient` and `HighRiskPatient` have a method named `describe()`.
   - Python chooses the correct version automatically.

This shows why OOP is useful in healthcare data projects.

```python
# Use the same patient 1001 example from the dataset
patient_1001 = Patient(1001, 45)
high_risk_1001 = HighRiskPatient(1001, 45, 'Diabetes')

print('--- Inheritance check ---')
print('HighRiskPatient is a subclass of Patient:', issubclass(HighRiskPatient, Patient))
print('high_risk_1001 has patient_id:', high_risk_1001.patient_id)
print('high_risk_1001 has age:', high_risk_1001.age)
print('high_risk_1001 also has condition:', high_risk_1001.condition)

print('\n--- Polymorphism check ---')
for obj in [patient_1001, high_risk_1001]:
    print(type(obj).__name__, '->', obj.describe())

print('\nWhy this matters:')
print('Inheritance lets HighRiskPatient reuse Patient fields.')
print('Polymorphism lets the same describe() method behave differently for each object.')
```

## What is inheritance? (definition)
Inheritance means a child class gets the features of a parent class.

## Why this matters in real code
A parent class holds the common rules, and the child class adds special behavior.

Example:
- `Patient` is the parent.
- `HighRiskPatient` and `FollowUpPatient` are children.

Both children reuse the basic `patient_id` and `age` logic from `Patient`, so we do not repeat code.

## What is polymorphism? (definition)
Polymorphism means the same method name can behave differently for different objects.

## Why this is powerful
The list `patients = [patient_a, patient_b, patient_c]` contains different objects, but this loop works for all of them:

for p in patients:
    print(p.describe())

Python checks the object type and chooses the correct `describe()` version automatically.

## What is polymorphism? (definition)
Polymorphism means the same method name can act differently depending on the object.

## Example of polymorphism
Both `Patient` and `HighRiskPatient` have a method called `describe()`, but they return different text:
- `Patient.describe()` gives a normal description.
- `HighRiskPatient.describe()` gives a more specific high-risk message.

So the same method name works in different forms for different classes.

```python
# Type hints example
# A type hint tells Python and the reader what kind of input and output to expect.
def calculate_cost(total_cost: float, visits: int) -> float:
    # total_cost should be a number like 2200.5
    # visits should be an integer like 6
    # The function returns a float result
    return total_cost / visits

# Call the function with real values from the dataset
print('Cost per visit:', calculate_cost(2200.5, 6))
```

```python
# Dataclass example
# A dataclass automatically creates a simple class with useful features like __init__ and __repr__.
from dataclasses import dataclass

@dataclass
class PatientRecord:
    patient_id: int      # patient id must be an integer
    age: int             # age must be an integer
    condition: str       # condition must be a text string

# Create a dataclass object
r1 = PatientRecord(1002, 32, 'Hypertension')
print('Dataclass object:', r1)
```
