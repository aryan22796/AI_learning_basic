# 1. Create a Simple Movie Dataset

This dataset represents users, movies, genres, ratings, and reviews.

In real companies, datasets are huge. But for learning, we start small.

````python
import pandas as pd

movie_data = [
    {"user": "Rahul", "age": 22, "movie": "Avengers", "genre": "Action", "rating": 5, "review": "Amazing action scenes and superheroes."},
    {"user": "Rahul", "age": 22, "movie": "Batman", "genre": "Action", "rating": 5, "review": "Dark action movie with a strong hero."},
    {"user": "Rahul", "age": 22, "movie": "Titanic", "genre": "Romance", "rating": 1, "review": "Too emotional and slow for me."},
    {"user": "Priya", "age": 25, "movie": "Titanic", "genre": "Romance", "rating": 5, "review": "Beautiful romantic story with emotional moments."},
    {"user": "Priya", "age": 25, "movie": "Notebook", "genre": "Romance", "rating": 5, "review": "Lovely emotional romantic drama."},
    {"user": "Amit", "age": 30, "movie": "Oppenheimer", "genre": "History", "rating": 5, "review": "Powerful historical drama with brilliant acting."},
    {"user": "Amit", "age": 30, "movie": "Dunkirk", "genre": "War", "rating": 4, "review": "Intense war movie with great direction."},
    {"user": "Neha", "age": 19, "movie": "Frozen", "genre": "Animation", "rating": 5, "review": "Fun animated movie with songs and family emotions."},
    {"user": "Neha", "age": 19, "movie": "Toy Story", "genre": "Animation", "rating": 5, "review": "Cute animated story with friendship and adventure."},
    {"user": "John", "age": 40, "movie": "Interstellar", "genre": "Sci-Fi", "rating": 5, "review": "Mind blowing science fiction and space travel."},
]

df = pd.DataFrame(movie_data)
df
````

# 2. What is Artificial Intelligence?

## Simple Definition

**Artificial Intelligence (AI)** means making machines perform tasks that normally require human intelligence.

Examples of human intelligence:

- Understanding language
- Recognizing faces
- Making decisions
- Solving problems
- Recommending something
- Predicting what can happen next

## In Our Movie Example

If a platform sees that Rahul likes action movies and recommends another action movie, it looks intelligent.

That is AI.

### Important Point

AI is the broad goal:

> Make computers behave intelligently.

## AI Example Using Our Dataset

Rahul gave high ratings to action movies.

So an AI system may recommend another action movie.

````python
rahul_movies = df[df["user"] == "Rahul"]
rahul_movies
````

````python
# Find Rahul's favorite genres based on high rating
rahul_favorites = rahul_movies[rahul_movies["rating"] >= 4]["genre"].value_counts()
rahul_favorites
````

````python
# Simple AI-style rule: recommend a movie from Rahul's favorite genre that he has not watched
favorite_genre = rahul_favorites.index[0]
watched_movies = set(rahul_movies["movie"])

recommendations = df[(df["genre"] == favorite_genre) & (~df["movie"].isin(watched_movies))]["movie"].unique()
recommendations
````

## Explanation

The system looked at Rahul's history.

It found:

- Rahul rated Avengers = 5
- Rahul rated Batman = 5
- Both are Action movies

So the system learned that Rahul may like Action movies.

Then it recommended other Action movies.

This is a simple example of AI behavior.

But this is still rule-based and very basic.

# 3. What is Machine Learning?

## Simple Definition

**Machine Learning (ML)** is a way of building AI systems where computers learn patterns from data instead of being manually programmed for every rule.

## Difference Between AI and ML

AI is the goal.

ML is one method to achieve that goal.

Example:

- AI goal: recommend movies intelligently
- ML method: learn from user ratings and past behavior

## In Our Movie Example

The system learns:

> Users who rate Action movies highly may prefer Action movies in the future.

This learning happens from data.

## ML Example: Predict Whether a User Likes a Movie

We will create a simple target column:

- rating >= 4 means user liked the movie
- rating < 4 means user did not like the movie

````python
df["liked"] = df["rating"].apply(lambda x: 1 if x >= 4 else 0)
df[["user", "movie", "genre", "rating", "liked"]]
````

Now we will use a very simple machine learning model.

Input feature:

- genre

Output prediction:

- liked or not liked

This is simplified for teaching. Real recommendation systems use many more features.

````python
from sklearn.preprocessing import OneHotEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

X = df[["genre"]]
y = df["liked"]

model = Pipeline(steps=[
    ("encoder", ColumnTransformer(
        transformers=[("genre_encoder", OneHotEncoder(handle_unknown="ignore"), ["genre"])]
    )),
    ("classifier", DecisionTreeClassifier(random_state=42))
])

model.fit(X, y)

# Predict whether a user may like an Action movie
sample = pd.DataFrame({"genre": ["Action"]})
prediction = model.predict(sample)

prediction
````

## Interpretation

The model gives a prediction:

- 1 means the user may like it
- 0 means the user may not like it

The important point is not model accuracy here.

The important point is:

> Machine Learning learns patterns from data.

We did not manually write every possible rule.

# 4. What is Deep Learning?

## Simple Definition

**Deep Learning** is a type of Machine Learning that uses neural networks with many layers.

It is useful when data is large and complex.

## Why Deep Learning?

Traditional ML works well for simple structured data like:

- age
- genre
- rating

But modern platforms have complex data:

- movie posters
- trailers
- subtitles
- user watch time
- search history
- reviews
- audio
- images
- text descriptions

Deep Learning can learn patterns from these complex data types.

## In Our Movie Example

Netflix-like systems do not only look at ratings.

They may analyze:

- the movie poster image
- the trailer video
- the review text
- how many minutes the user watched
- whether user skipped the movie
- which actor user likes

This requires more powerful models.

## Deep Learning Concept Diagram

```text
Input Data
    |
    |-- User age
    |-- Movie genre
    |-- Rating history
    |-- Review text
    |-- Poster image
    |-- Watch time
    |
    v
Neural Network Layers
    |
    v
Prediction / Recommendation
```

Deep Learning means the model learns multiple levels of features.

Example from movie poster:

- Layer 1 may detect colors and edges
- Layer 2 may detect faces or objects
- Layer 3 may detect action scene or romantic scene
- Final layer predicts genre or recommendation

````python
# Simple neural-network-like example using sklearn MLPClassifier
# This is not a full deep learning model, but useful for beginner demonstration.

from sklearn.neural_network import MLPClassifier

X = df[["genre", "age"]]
y = df["liked"]

nn_model = Pipeline(steps=[
    ("encoder", ColumnTransformer(
        transformers=[
            ("genre_encoder", OneHotEncoder(handle_unknown="ignore"), ["genre"]),
            ("age_passthrough", "passthrough", ["age"])
        ]
    )),
    ("classifier", MLPClassifier(hidden_layer_sizes=(8, 4), max_iter=1000, random_state=42))
])

nn_model.fit(X, y)

sample = pd.DataFrame({"genre": ["Sci-Fi"], "age": [35]})
nn_prediction = nn_model.predict(sample)

nn_prediction
````

## Important Teaching Note

This small neural network is only for demonstration.

Real deep learning models use:

- millions or billions of data points
- GPUs
- large neural networks
- huge training time

But the basic idea is the same:

> Deep Learning uses layered neural networks to learn complex patterns.

# 5. What is Generative AI?

## Simple Definition

**Generative AI** is AI that can create new content.

It can generate:

- text
- code
- images
- audio
- video
- music
- summaries
- emails
- stories

## Difference from Traditional AI

Traditional AI may classify or predict.

Generative AI creates something new.

## In Our Movie Example

Recommendation AI:

> You may like Avengers.

Generative AI:

> Write a new superhero movie story for Rahul.

That is the difference.

## Generative AI Example

Prompt:

```text
Write a short movie story about an Indian superhero who protects Mumbai.
```

Possible output:

```text
Title: The Guardian of Mumbai

A young engineer discovers a hidden power after saving people during a storm.
He uses technology and courage to protect the city from a dangerous cyber-criminal.
```

This content did not exist in the dataset.

The model generated it.

````python
# Simple template-based generation demo
# This is NOT a real LLM. It is only to explain the idea of generation.

user_name = "Rahul"
favorite_genre = "Action"
city = "Mumbai"

story = f"""
Title: The {favorite_genre} Hero of {city}

{user_name} is an ordinary person who discovers a hidden power.
When danger comes to {city}, he decides to protect the people.
With courage, discipline, and intelligence, he becomes the city's new hero.
"""

print(story)
````

## Teaching Note

The above code is template-based generation, not real Generative AI.

A real Generative AI model does not use a fixed template only.

It learns language patterns from huge data and generates flexible responses token by token.

This is where LLMs come in.
